```python
import pandas as pd
import numpy as np
```

```python
file_path = "gather_boost.csv"
gather_boost_df = pd.read_csv(file_path)
gather_boost_df.head()
```

```python
gather_boost_df = gather_boost_df[['industry_type_subpart', 'total_emissions_2020', 'total_emissions_2019', 'total_emissions_2018',
       'total_emissions_2017', 'total_emissions_2016']]
gather_boost_df
```

```python
avg_gather_boost_emissions2020 = gather_boost_df["total_emissions_2020"].mean()
avg_gather_boost_emissions2019 = gather_boost_df["total_emissions_2019"].mean()
avg_gather_boost_emissions2018 = gather_boost_df["total_emissions_2018"].mean()
avg_gather_boost_emissions2017 = gather_boost_df["total_emissions_2017"].mean()
avg_gather_boost_emissions2016 = gather_boost_df["total_emissions_2016"].mean()
```

```python
data = {'year': ['2020', '2019', '2018', '2017', '2016'],
        'avg_emissions': [avg_gather_boost_emissions2020, avg_gather_boost_emissions2019, avg_gather_boost_emissions2018, avg_gather_boost_emissions2017,
                         avg_gather_boost_emissions2016]}
yearly_avg_df = pd.DataFrame(data)
yearly_avg_df
```

```python
yearly_avg_df.sort_values(by=['year'], inplace=True)
yearly_avg_df
```

```python
import matplotlib.pyplot as plt
```

```python
yearly_avg_df.plot.bar(x="year", y="avg_emissions")
```

```python
year = list(data['year'])
avg_emissions = list(data['avg_emissions'])

 
# creating the bar plot
plt.bar(year, avg_emissions,
        width = 0.4)
 
plt.xlabel("Year")
plt.ylabel("GHG Emissions")
plt.title("Average Gathering & Boosting GHG Emissions per Year")
plt.show()
```
