```python
import pandas as pd
import numpy as np
```

```python
onshore_oil_gas = pd.read_excel (r'resources/ghgp_data_by_year.xlsx', sheet_name='Onshore Oil & Gas Prod.', header=3)
onshore_oil_gas.head(10)
```

```python
df = onshore_oil_gas.drop(columns=['FRS Id',
                              'Reported Address',
                              'Reported County', 
                             ])
df.columns
```

```python
# Rename the columns
col_names = {'Facility Id': 'facility_id', 
             'Facility Name': 'facility_name', 
             'Basin': 'basin', 
             'Reported City': 'city', 
             'Reported State': 'state', 
             'Reported Zip Code': 'zip_code',
             'Reported Latitude': 'latitude',
             'Reported Longitude': 'longitude',
             'Primary NAICS Code': 'primary_naics_code',
       'Latest Reported Industry Type (subparts)': 'industry_type_subpart',
       '2020 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2020',
       '2019 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2019',
       '2018 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2018',
       '2017 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2017',
       '2016 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2016',
       '2015 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2015',
       '2014 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2014',
       '2013 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2013',
       '2012 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2012',
       '2011 Total reported emissions from Onshore Oil & Gas Production ': 'total_emissions_2011'}
df.rename(columns=col_names, inplace=True)
```

```python
basins = df['basin'].str.split(pat='-', expand=True)
```

```python
df[['basin_num','basin_name','extra']] = basins
#gather_boost_df = gather_boost_df.drop(['extra'])
```

```python
df = df.drop(['basin','extra'], axis=1)
df
```

```python
df = df.dropna()
df
```

```python
## Export to CSV without the index
df.to_csv('basins.csv', index=False) 
```
