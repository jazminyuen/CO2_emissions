```python
# Import dependencies
import pandas as pd
```

```python
# Import excel file and create DataFrame
file_to_load = "ghgp_data_by_year.xlsx"
tr_pipe_df = pd.read_excel(file_to_load, sheet_name="Transmission Pipelines", header=3)
tr_pipe_df.head(10)
```

```python
tr_pipe_df.columns
```

```python
# Drop unwanted columns
df = tr_pipe_df[['Facility Id', 'Facility Name', 'State where Emissions Occur',
                 'Primary NAICS Code',
                 'Latest Reported Industry Type (subparts)',
       '2020 Total reported direct emissions from Transmission Pipelines',
       '2019 Total reported direct emissions from Transmission Pipelines',
       '2018 Total reported direct emissions from Transmission Pipelines',
       '2017 Total reported direct emissions from Transmission Pipelines',
       '2016 Total reported direct emissions from Transmission Pipelines']]
df.columns
```

```python
# Rename the columns
col_names = {'Facility Id': 'facility_id', 'Facility Name': 'facility_name', 'State where Emissions Occur': 'state',
       'Primary NAICS Code', 'Latest Reported Industry Type (subparts)',
       '2020 Total reported direct emissions from Transmission Pipelines',
       '2019 Total reported direct emissions from Transmission Pipelines',
       '2018 Total reported direct emissions from Transmission Pipelines',
       '2017 Total reported direct emissions from Transmission Pipelines',
       '2016 Total reported direct emissions from Transmission Pipelines'}
df = df.rename(columns=col_names)
```

```python
# Drop null values
df_not_null = df.dropna()
df_not_null.count()
```

```python
# Export to CSV without the index
df_not_null.to_csv('transmission_pipelines.csv', index=False, sep=',', header=True) 
```
