```python
import pandas as pd
import numpy as np
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
direct_emitters_df = pd.read_excel (r'resources/ghgp_data_by_year.xlsx', sheet_name='Direct Emitters', header=3)
direct_emitters_df.head(10)
```

```python
direct_emitters_df.columns
```

```python
direct_emitters_df = direct_emitters_df[['Facility Id', 'Facility Name', 'City', 'State', 
        'Zip Code', 'Latitude', 'Longitude',
        'Primary NAICS Code','Latest Reported Industry Type (subparts)',
       'Latest Reported Industry Type (sectors)',
       '2020 Total reported direct emissions',
       '2019 Total reported direct emissions',
       '2018 Total reported direct emissions',
       '2017 Total reported direct emissions',
       '2016 Total reported direct emissions',
       '2015 Total reported direct emissions',
       '2014 Total reported direct emissions',
       '2013 Total reported direct emissions',
       '2012 Total reported direct emissions',
       '2011 Total reported direct emissions' ]]
direct_emitters_df
```

```python
direct_emitters_df = direct_emitters_df.dropna()
direct_emitters_df
```

```python
direct_emitters_df.rename(columns = {'Facility Id':'facility_id',
                              'Facility Name':'facility_name', 'City':'city',
                                     'State':'state','Zip Code':'zip_code', 'Latitude': 'latitude', 'Longitude': 'longitude',
                                     'Primary NAICS Code':'primary_naics_code',
                                    'Latest Reported Industry Type (subparts)':'industry_type_subpart',
                                     'Latest Reported Industry Type (sectors)':'industry_type_sector',
                                     '2020 Total reported direct emissions':'total_emissions_2020',
                                     '2019 Total reported direct emissions':'total_emissions_2019',
                                     '2018 Total reported direct emissions':'total_emissions_2018',
                                     '2017 Total reported direct emissions':'total_emissions_2017',
                                     '2016 Total reported direct emissions':'total_emissions_2016',
                                     '2015 Total reported direct emissions':'total_emissions_2015',
                                     '2014 Total reported direct emissions':'total_emissions_2014',
                                     '2013 Total reported direct emissions':'total_emissions_2013',
                                     '2012 Total reported direct emissions':'total_emissions_2012',
                                     '2011 Total reported direct emissions':'total_emissions_2011'},
                                     inplace = True)
direct_emitters_df
```

```python
direct_emitters_df['total_emissions'] = direct_emitters_df.sum(axis=1, numeric_only=True)
direct_emitters_df
```

```python
direct_emitters_total = direct_emitters_df[['industry_type_sector', 'total_emissions']]
direct_emitters_total
```

```python
industry_counts = direct_emitters_total.industry_type_sector.value_counts()
industry_counts
```

```python
# Determine which values to replace
replace_industry = list(industry_counts[industry_counts < 100].index)

# Replace in DataFrame
for industry in replace_industry:
    direct_emitters_total.industry_type_sector = direct_emitters_total.industry_type_sector.replace(industry,"Other")


# Check to make sure binning was successful
direct_emitters_total.industry_type_sector.value_counts()
```

```python
industry_num = {
   "Other": 1,
   "Power Plants": 2,
   "Waste": 3,
   "Petroleum and Natural Gas Systems": 4,
   "Minerals": 5,
   "Chemicals": 6,
   "Metals": 7,
   "Pulp and Paper": 8,
}
```

```python
direct_emitters_total["industry_type_sector"] = direct_emitters_total["industry_type_sector"].apply(lambda x: industry_num[x])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(direct_emitters_total)
scaled_data = scaler.transform(direct_emitters_total)

transformed_scaled_data = pd.DataFrame(scaled_data, columns=direct_emitters_total.columns)
transformed_scaled_data.head()
```

```python
# Find the best value for K
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(direct_emitters_total)
    inertia.append(km.inertia_)

# Create the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# Initializing model with K = 3 
model = KMeans(n_clusters=3, random_state=5)
model
```

```python
# Fitting model
model.fit(direct_emitters_total)
```

```python
# Get the predictions
predictions = model.predict(direct_emitters_total)
print(predictions)
```

```python
# Add a new class column to the direct_emitters_ml
direct_emitters_total["class"] = model.labels_
direct_emitters_total.head()
```

```python
direct_emitters_total.drop('industry_type_sector', axis=1, inplace=True)
direct_emitters_total
```

```python
merged_df = direct_emitters_total.merge(direct_emitters_df, how='right', on='total_emissions')
merged_df
```
