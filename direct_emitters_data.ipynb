```python
import pandas as pd
```

```python
direct_emitters_df = pd.read_excel (r'resources/ghgp_data_by_year.xlsx', sheet_name='Direct Emitters', header=3)
direct_emitters_df.head(10)
```

```python
direct_emitters_df.columns
```

```python
direct_emitters_df = direct_emitters_df[['Facility Id', 'Facility Name', 'City', 'State', 'Zip Code', 'Primary NAICS Code','Latest Reported Industry Type (subparts)',
       'Latest Reported Industry Type (sectors)',
       '2020 Total reported direct emissions',
       '2019 Total reported direct emissions',
       '2018 Total reported direct emissions',
       '2017 Total reported direct emissions',
       '2016 Total reported direct emissions',
       '2015 Total reported direct emissions',
       '2014 Total reported direct emissions',
       '2013 Total reported direct emissions',
       '2012 Total reported direct emissions',
       '2011 Total reported direct emissions' ]]
direct_emitters_df
```

```python
direct_emitters_df = direct_emitters_df.dropna()
direct_emitters_df
```

```python
direct_emitters_df.rename(columns = {'Facility Id':'facility_id',
                              'Facility Name':'facility_name', 'City':'city',
                                     'State':'state','Zip Code':'zip_code','Primary NAICS Code':'primary_naics_code',
                                    'Latest Reported Industry Type (subparts)':'industry_type_subpart',
                                     'Latest Reported Industry Type (sectors)':'industry_type_sector',
                                     '2020 Total reported direct emissions':'total_emissions_2020',
                                     '2019 Total reported direct emissions':'total_emissions_2019',
                                     '2018 Total reported direct emissions':'total_emissions_2018',
                                     '2017 Total reported direct emissions':'total_emissions_2017',
                                     '2016 Total reported direct emissions':'total_emissions_2016',
                                     '2015 Total reported direct emissions':'total_emissions_2015',
                                     '2014 Total reported direct emissions':'total_emissions_2014',
                                     '2013 Total reported direct emissions':'total_emissions_2013',
                                     '2012 Total reported direct emissions':'total_emissions_2012',
                                     '2011 Total reported direct emissions':'total_emissions_2011'},
                                     inplace = True)
direct_emitters_df
```

```python
direct_emitters_df.to_csv('resources/Direct_Emitters_cleaned.csv', index=False, sep=',', header=True)
```
