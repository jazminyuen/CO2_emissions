```python
import pandas as pd
import numpy as np
```

```python
file_path = "onshore_oil_gas.csv"
onshore_oil_df = pd.read_csv(file_path)
onshore_oil_df.head()
```

```python
onshore_oil_df = onshore_oil_df[['industry_type_subpart', 'total_emissions_2020', 'total_emissions_2019', 'total_emissions_2018',
       'total_emissions_2017', 'total_emissions_2016', 'total_emissions_2015',
       'total_emissions_2014', 'total_emissions_2013', 'total_emissions_2012',
       'total_emissions_2011']]
onshore_oil_df
```

```python
avg_onshore_oil_emissions2020 = onshore_oil_df["total_emissions_2020"].mean()
avg_onshore_oil_emissions2019 = onshore_oil_df["total_emissions_2019"].mean()
avg_onshore_oil_emissions2018 = onshore_oil_df["total_emissions_2018"].mean()
avg_onshore_oil_emissions2017 = onshore_oil_df["total_emissions_2017"].mean()
avg_onshore_oil_emissions2016 = onshore_oil_df["total_emissions_2016"].mean()
avg_onshore_oil_emissions2015 = onshore_oil_df["total_emissions_2015"].mean()
avg_onshore_oil_emissions2014 = onshore_oil_df["total_emissions_2014"].mean()
avg_onshore_oil_emissions2013 = onshore_oil_df["total_emissions_2013"].mean()
avg_onshore_oil_emissions2012 = onshore_oil_df["total_emissions_2012"].mean()
avg_onshore_oil_emissions2011 = onshore_oil_df["total_emissions_2011"].mean()
```

```python
data = {'year': ['2020', '2019', '2018', '2017', '2016', '2015', '2014', '2013', '2012', '2011'],
        'avg_emissions': [avg_onshore_oil_emissions2020, avg_onshore_oil_emissions2019, avg_onshore_oil_emissions2018, avg_onshore_oil_emissions2017,
                         avg_onshore_oil_emissions2016, avg_onshore_oil_emissions2015, avg_onshore_oil_emissions2014, avg_onshore_oil_emissions2013,
                         avg_onshore_oil_emissions2012, avg_onshore_oil_emissions2011]}
yearly_avg_df = pd.DataFrame(data)
yearly_avg_df
```

```python
yearly_avg_df.sort_values(by=['year'], inplace=True)
yearly_avg_df
```

```python
import matplotlib.pyplot as plt
```

```python
yearly_avg_df.plot.bar(x="year", y="avg_emissions")
```

```python
year = list(data['year'])
avg_emissions = list(data['avg_emissions'])

 
# creating the bar plot
plt.bar(year, avg_emissions,
        width = 0.4)
 
plt.xlabel("Year")
plt.ylabel("GHG Emissions")
plt.title("Average Onshore Oil & Gas GHG Emissions per Year")
plt.show()
```
