```python
import pandas as pd
import plotly.express as px
import hvplot.pandas
from sklearn.cluster import KMeans
```

```python
file_path = "Direct_Emitters_cleaned.csv"
direct_emitters_df = pd.read_csv(file_path, index_col=False)
direct_emitters_df.head()
```

```python
direct_emitters_df.columns
```

```python
direct_emitters_df = direct_emitters_df[['industry_type_sector', 'total_emissions_2020', 'total_emissions_2019', 'total_emissions_2018',
       'total_emissions_2017', 'total_emissions_2016', 'total_emissions_2015',
       'total_emissions_2014', 'total_emissions_2013', 'total_emissions_2012',
       'total_emissions_2011']]
direct_emitters_df
```

```python
print(f'Duplicated Entries: {direct_emitters_df.duplicated().sum()}')
```

```python
# direct_emitters_df = direct_emitters_df.set_index('industry_type_sector')
# direct_emitters_df
```

```python
direct_emitters_df['total_emissions'] = direct_emitters_df.sum(axis=1, numeric_only=True)
direct_emitters_df
```

```python
direct_emitters_total = direct_emitters_df[['industry_type_sector', 'total_emissions']]
direct_emitters_total
```

```python
industry_counts = direct_emitters_total.industry_type_sector.value_counts()
industry_counts
```

```python
# Determine which values to replace
replace_industry = list(industry_counts[industry_counts < 100].index)

# Replace in DataFrame
for industry in replace_industry:
    direct_emitters_total.industry_type_sector = direct_emitters_total.industry_type_sector.replace(industry,"Other")


# Check to make sure binning was successful
direct_emitters_total.industry_type_sector.value_counts()
```

```python
industry_num = {
   "Other": 1,
   "Power Plants": 2,
   "Waste": 3,
   "Petroleum and Natural Gas Systems": 4,
   "Minerals": 5,
   "Chemicals": 6,
   "Metals": 7,
   "Pulp and Paper": 8,
}
```

```python
direct_emitters_total["industry_type_sector"] = direct_emitters_total["industry_type_sector"].apply(lambda x: industry_num[x])
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(direct_emitters_total)
scaled_data = scaler.transform(direct_emitters_total)

transformed_scaled_data = pd.DataFrame(scaled_data, columns=direct_emitters_total.columns)
transformed_scaled_data.head()
```

```python
# Find the best value for K
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(direct_emitters_total)
    inertia.append(km.inertia_)

# Create the elbow curve
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# Initializing model with K = 3 
model = KMeans(n_clusters=3, random_state=5)
model
```

```python
# Fitting model
model.fit(direct_emitters_total)
```

```python
# Get the predictions
predictions = model.predict(direct_emitters_total)
print(predictions)
```

```python
# Add a new class column to the direct_emitters_ml
direct_emitters_total["class"] = model.labels_
direct_emitters_total.head()
```

```python
# Create a scatterplot of direct_emitters_ml
direct_emitters_total.hvplot.scatter(x="total_emissions", y="industry_type_sector", by="class")
```

```python
# Plotting the clusters with three features
fig = px.scatter_3d(direct_emitters_total, x="total_emissions", y="industry_type_sector", z="class", color="class", symbol="class", width=800)
fig.update_layout(legend=dict(x=0,y=1))
fig.show()
```
